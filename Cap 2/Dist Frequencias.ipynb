# Medidas de Frequência

```python
df_grau_instrucao = df['grau_instrucao']
freq_grau_instrucao = df_grau_instrucao.value_counts()
proporcao_grau_instrucao = round(freq_grau_instrucao / sum(freq_grau_instrucao), 3)
porcentagem_grau_instrucao = round(proporcao_grau_instrucao * 100, 2)
```

```python
dist_grau_instrucao = pd.DataFrame(
    {
      'Frequência': freq_grau_instrucao, 
      'Proporção': proporcao_grau_instrucao, 
      'Porcentagem (%)': porcentagem_grau_instrucao
      }
)
dist_grau_instrucao
```

```python

"""
Distribuição frequência por faixa salárial

"""

faixas_salarios = [4 , 8,  12,  16, 20, 24]


labels = ['4', '8', '12', '16', '20']
frequencia = pd.value_counts(
    pd.cut(x= df['salario'],
       bins= faixas_salarios,
       labels= labels,
       include_lowest=True))
frequencia
```

```python
''' 
Porcentagem de acordo com a distribuição de frequência por faixa salárial

'''

percentual = pd.value_counts(
    pd.cut(x= df['salario'],
           bins= faixas_salarios,
           labels= labels,
           include_lowest= True),
    normalize = True
) *100
percentual
```

```python
dist_faixa_salarial = pd.DataFrame(
    {
        'Quantidade por Salario minimo': frequencia,
        'Porcentagem (%)': percentual
    }
)
dist_faixa_salarial.sort_values('salario', ascending=True)
```

```python
'''
Distribuição de Frequência de funcionários por região

'''

contagem_regiao = df['reg_procedencia'].value_counts()
freq_por_regiao = contagem_regiao / round(sum(contagem_regiao), 3)
porcentagem_por_regiao = round(freq_por_regiao * 100, 2)
```

```python
df_dist_freq_regiao = pd.DataFrame(
    {
        'Região': contagem_regiao,
        'Frequência': freq_por_regiao,
        'Porcentagem(%)': porcentagem_por_regiao
    }
)
df_dist_freq_regiao.sort_values('Frequência', ascending= True)
```

```python
"""
Distribuição por estado civil

"""
cont_estado_civil = df.value_counts('estado_civil')
freq_estado_civil = round(cont_estado_civil / sum(cont_estado_civil), 2)
porcentagem_estado_civil = round(freq_estado_civil * 100, 2)
```

```python
dist_freq_estado_civil = pd.DataFrame(
    {
        'Estado civil': cont_estado_civil, 
        'Frequência': freq_estado_civil,
        'Porcentagem(%)': porcentagem_estado_civil,
    }
)
dist_freq_estado_civil
```

```python
'''
Distribuição de Frequência da quantidade de filhos de funcionários casados 

'''
casados = df.loc[df['estado_civil'] == 'casado']
casados.head()

```

```python
funcio_casados_n_filhos = pd.crosstab(casados['estado_civil'],
            casados['n_filhos'])
funcio_casados_n_filhos
freq_filhos_por_casado = round(funcio_casados_n_filhos / sum(casados.value_counts()), 2)
porcentagem_filhos_por_casado = round(freq_filhos_por_casado * 100, 2)

porcentagem_filhos_por_casado
```

```python
dist_freq_casado_filhos = pd.DataFrame(
    {
        'Funcionários Casados': funcio_casados_n_filhos,
        'Frequência': freq_filhos_por_casado,
        'Porcentagem (%)': porcentagem_filhos_por_casado
    }
)
```

```python
casados_agg_filhos = casados.groupby( 'n_filhos').value_counts().reset_index()
casados_agg_filhos.sort_values('N',ascending=True).head(3)
```

```python
contagem_filhos = casados_agg_filhos['n_filhos'].value_counts()
contagem_filhos
freq_filhos_por_casado = round(contagem_filhos / sum(contagem_filhos), 2)
porcentagem_filhos_por_casado = round(freq_filhos_por_casado * 100, 2)
```

```python
df_freq_quantidade_casados_filhos = pd.DataFrame({
    'Estado Civil': 'Casado',
    'Número de filhos': contagem_filhos,
    'Frequência': freq_filhos_por_casado,
    'Porcentagem (%)': porcentagem_filhos_por_casado
})
df_freq_quantidade_casados_filhos
```
